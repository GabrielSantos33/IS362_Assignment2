##### The sip function takes in iterables as arguments and returns an iterator. This iterator generates a series of tuples containing elements from each iterable. zip() can accept any type of iterable, such as files, lists, tuples, dictionaries, sets, and so on.

```python
tup1 = ('a', 'b', 'c' , 'd', 'e')
tup2 = (10, 20, 30, 40, 50, 60)

set1 = zip(tup1, tup2)
```

##### Convert a set into a dictionary.

```python
newdict = dict(set1)
print (newdict)
```

##### This is a interesting way to work with data frames using pandas. We import pandas module then the Series.

##### Series is a type of list in pandas which can take integer values, string values, double values and more. Series can only contain single list with index.

##### Dataframes can be made of more than one series or we can say that a dataframe is a collection of series that can be used to analyse the data.

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

##### First we create a series (array)

```python
grades = Series([90, 80, 98, 79, 100, 82, 91])
print(grades)
```

##### Another series

```python
students = Series(['John', 'Edward', 'Gabriel', 'Michael', 'Angela', 'Peter', 'Tara'])
print(students)
```

##### And another...

```python
dates = pd.date_range('2021-02-01','2021-02-07')
print(dates)
```

##### Now we create a Dataframe StudentDf convining the 3 series created previously.

```python
StudentDf = DataFrame({'Student':students,
                      'Grades':grades,
                      'Date':dates})
print(StudentDf)
```
